<h1> Графики </h1>

<h4> 1. Распределение целевой переменной </h4>

```python
sns.countplot(x='Exited', data = data_train)
plt.title("Распределение оттока клиентов")
plt.show()
```

<h4>2. Визуализация возраста </h4>

```python
sns.histplot(data_train['Age'], bins = 30, kde = True)
plt.title("Распределение возрастов клиентов")
plt.show()
```

<h4>3. Визуализация баланса</h4>

```python
sns.histplot(data = data_train, x = 'Balance', kde = False)
plt.title('Распределение баланса')
plt.show()
```

<h4>4. Визуализация корелляции с Exited</h4>

```python
data_drop = data_train.drop(['id', 'CustomerId', 'Surname', 'Geography'], axis = 1).copy()
data_dummy = pd.get_dummies(data_drop)
data_corr  = data_dummy.corr()

plt.figure(figsize = (14, 8))
sns.heatmap(data_corr, cmap = 'coolwarm', center = 0, annot = True)
plt.title('Матрица корелляции')
plt.show()
```
